# Titanic Basic Analysis

This notebook provides a basic analysis of the Titanic dataset focusing on the following aspects:

1. **Age Distribution**: Visualization of age distribution among Titanic passengers.
2. **Gender Class Distribution**: Examination of gender-based distribution across passenger classes.
3. **Passenger Class Analysis**: Analysis of survival rates across different passenger classes.

## Libraries Import

Import necessary libraries for the analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style='whitegrid')
```

## Data Loading

Load the cleaned Titanic dataset. Handle errors in file loading to ensure smooth execution.

```python
try:
    titanic_df = pd.read_csv('titanic_cleaned.csv')
    print('Data loaded successfully.')
except FileNotFoundError:
    print('Error: File not found. Please check the file path.')
except Exception as e:
    print(f'An error occurred: {e}')
```

## Age Distribution

Visualize the distribution of passenger ages using histograms with overlaid KDE plots.

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['age'].dropna(), kde=True, color='blue', bins=30)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
